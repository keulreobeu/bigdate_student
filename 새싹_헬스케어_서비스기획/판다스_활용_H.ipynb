### 데이터간의 연산처리
DataFrame에서 연산처리 또한 특정 변수를 의미하는 컬럼단위에서 연산 처리가 주로 발생하기 때문에 먼저 연산처리에 필요한 컬럼을 선택하여 시리즈로 가져옵니다.

시리즈에서 연산자를 사용하면 데이터간의 일괄연산을 수행할 수 있습니다.

- `시리즈 연산자 단일데이터`: 시리즈내 모든 데이터에 단일데이터를 일괄연산
- `시리즈 연산자 시리즈`: 두 시리즈간 동일한 인덱스끼리 일괄연산


```python
# 예제 데이터 불러오기
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원.csv',
                  encoding='CP949',
                  index_col=2,
                  header=1)

df_gp
```

#### 연산자를 활용한 일괄연산

연산 결과를 `DataFrame`에 반영하기 위해선 연산 결과 얻은 시리즈를 다시 동일 컬럼에 재할당해야만 합니다.   

```python
# DataFrame에 연산자 적용시 일곽적용
df_gp['보험자부담금'] = df_gp['보험자부담금'] / 10000
df_gp['보험자부담금'] = df_gp['보험자부담금'].astype('int')
df_gp
```

기존 `DataFrame`에 연산 결과를 적용할 때 새로운 컬럼 이름으로 결과 시리즈를 할당하면 새로운 컬럼으로 추가됩니다.

```python
# 시리즈 끼리의 일관연산(같은 인덱스끼리 연산)
df_gp['요양급여비평균'] = df_gp['요양급여비용총액'] / df_gp['환자수']
df_gp['요양급여비평균'] = df_gp['요양급여비평균'].astype('int')
df_gp
```

연산자를 활용하는 방법말고도 다양한 연산 함수를 사용할 수 있습니다.

## 데이터 집계
데이터분석에 있어 가장 많이 활용되는 분석 방식은 데이터를 그룹화 하여 통합적인 통계수치를 나타내는 집계입니다.
pandas에는 합계, 평균과 같은 다양한 집계 함수를 제공하며, 기본적으로 데이터그룹 단위인 컬럼(시리즈)에서 활용 가능합니다.

**pandas 집계 함수**
- `size` : 관측치 개수
- `count` : 결측치를 제외한 관측치 개수
- `max`: 최댓값
- `min`: 최솟값
- `mean` : 평균
- `median` : 중앙값
- `sum` : 총합
- `std` : 표준편차
- `var` : 분산
- `quantile`: 샘플 분위수(% 값)
- `nunique`: 고유항목수
- `idxmax`: 최댓값 인덱스
- `agg`: 여러개의 집계


### 집계 함수 사용하기

집계함수는 통계 수치를 나타내기 때문에 주로 수치형 타입의 `Series`에서 활용 가능합니다.  
몇몇 집계 함수는 `DataFrame`에서도 사용 가능하며 요구되는 데이터타입이 전부 맞추어져야 합니다.


```python
# 예제 데이터 불러오기
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원_진료비.csv')
div = df_gp[['요양급여비용총액', '보험자부담금']] / 10000
df_gp[['요양급여비용총액', '보험자부담금']] = div.astype('int64')
df_gp
```

```python
# 시리즈에서 mean() 함수를 활용하여 평균 계산
df_gp['환자수'].mean()
```

`DataFrame`에서 집계함수 사용시 각 컬럼별로 집계가 이루어져 컬럼이름과 각 집계 값이 매핑 되는 시리즈를 반환합니다.

```python
# 데이터프레임에서 sum()으로 합계 계산
df_gp[['환자수','명세서청구건수']].sum()
```

```python
# var()으로 분산 계산
df_gp['보험자부담금'].var()
```

```python
# std()으로 표준편차계산
df_gp['보험자부담금'].std()
```

```python
# idxmax()로 가장큰 값의 인덱스 출력
df_gp['보험자부담금'].idxmax()
```

```python
# agg()로 여러개의 집계함수를 적용
df_gp['보험자부담금'].agg(['max','mean'])
```

```python
# count()으로 개수 계산 (결측치 제외)
df_gp['보험자부담금'].count()
```

```python
# value_counts()으로 항목 개수를 계산
df_gp['시군구'].value_counts()
```

### *describe()* 함수를 활용한 집계 요약표 생성

```python
# 각 컬럼의 통계 수치를 나타내는 통계 표 생성
df_gp[['환자수','요양급여비용총액','보험자부담금']].describe()
```

### *groupby()* 함수를 활용한 그룹별 통계 수치
`groupby()`함수를 활용하면 단순 컬럼간의 집계가 아닌 내부데이터중 동일한 데이터를 그룹화 하여 그룹에 대한 집계를 수행할 수 있습니다.


<center><img src="https://drive.google.com/thumbnail?id=1CDUfZontPTaA28AVlGIbqmHerdutGtAm&sz=w1000" width="800"/></center>


[참고](https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html)

</center>

    groupby(by='그룹화컬럼')['집계컬럼'].통계함수()

- **그룹화 (Splitting)**: 범주형 컬럼을 지정하여 컬럼내 동일한 데이터를 그룹화
- **인덱싱(Indexing)**: 지정한 그룹에 대해 집계가 이루어질 컬럼 선택
- **적용 (Applying)**: 각 그룹에 대해 집계 함수(예: `sum`, `mean`, `count` 등)나 변환 함수를 적용
- **결합 (Combining)**: 적용된 결과를 하나의 데이터프레임으로 반환
  

```python
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원_진료비.csv')
div = df_gp[['요양급여비용총액', '보험자부담금']] / 10000
df_gp[['요양급여비용총액', '보험자부담금']] = div.astype('int64')
```

그룹화가 진행될 컬럼은 `as_index` 옵션에 따라 그룹화 결과표의 인덱스가 될지 컬럼이 될지가 설정며되, 이에따라 결과가 시리즈가 될수도 있습니다.

- `as_index = True`: 그룹화 컬럼이 인덱스로 설정(기본값)
- `as_index = False`: 그룹화 컬럼이 컬럼으로 유지


```python
# 그룹별 평균 계산
group = df_gp.groupby(by='시도')['요양급여비용총액'].sum()
group
```

> 표현할 컬럼이 집계 컬럼 1개 이므로 시리즈로 표현

```python
# as_index 활용
group = df_gp.groupby('시도', as_index=False)['요양급여비용총액'].sum()
group
```

> 표현할 컬럼이 그룹화 컬럼과 집계 컬럼 2개 이상으로 DataFrame으로 표현

`[]` 인덱싱 연산에 집계컬럼을 여러개 리스트로 입력 하여 컬럼별 일괄 집계 가능합니다.

```python
# 여러개의 컬럼으로 집계
df_gp.groupby(by='시도')[['환자수', '요양급여비용총액']].sum()
```

그룹화 컬럼을 여러개로 지정하면 멀티인덱스 형식을 가진 DataFrame결과가 반환됩니다.

```python
# 여러개 컬럼을 그룹화 하여 평균 계산
df_gp.groupby(['시도', '진료년도'])[['환자수','요양급여비용총액']].sum()
```

`agg()` 함수를 활용하여 여러 집계 함수를 일괄 적용 가능합니다.

```python
# agg()를 활용한 그룹별 여러 집계 함수 적용
df_gp.groupby('시도', as_index=False)['환자수'].agg(['min','max','mean'])
```





### *pivot_table()* 을 활용한 피벗테이블 생성
`pivot_table()` 또한 데이터를 그룹화 하여 집계하는 방법중 하나입니다. `groupby()`와는 다르게 두개의 그룹 목록으로 부터 테이블 형식으로 집계를 나타낸다는 특징이 있습니다.  

<center><img src="https://drive.google.com/thumbnail?id=1YuOli4S5qBkmKqSXZMSPXNcHPexh0Pxb&sz=w1000" width="600"/></center>

[참고](https://pandas.pydata.org/docs/getting_started/intro_tutorials/07_reshape_table_layout.html)

- **그룹화**: 두개의 컬럼을 선택하여 행(row)과 열(columns)로 각각 그룹화
- **집계**: 하나의 집계 컬럼에 대해 두그룹에 포함되는 데이터를 집계
- **재구조화**: 계산된 결과를 새로운 행과 열 인덱스를 가진 표 형태로 반환


```python
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원_진료비.csv')
div = df_gp[['요양급여비용총액', '보험자부담금']] / 10000
df_gp[['요양급여비용총액', '보험자부담금']] = div.astype('int64')
```

- `index`: 그룹화 하여 인덱스가될 기존 컬럼명     
- `columns`: 그룹화 하여 컬럼이될 기존 컬럼명     
- `values`: 그룹별 집계가될 기존 컬럼명     
- `aggfunc`: 집계함수 이름

```python
# pivot_table() 활용
pivot = pd.pivot_table(df_gp, index='시도',
                       columns='진료년도',
                       values='환자수',
                       aggfunc='sum')
pivot
```

`aggfunc` 인자에는 pandas에서 제공하는 집계함수의 이름(문자열) 외에도 외부에서 구현된 함수 이름을 입력하여 활용 가능합니다.

```python
# 외부 통계함수 지원
import numpy as np

pivot = pd.pivot_table(df_gp, index='시도',
                       columns='진료년도',
                       values='환자수',
                       aggfunc=np.mean)
pivot
```

`groupby()`와 결과의 형태는 다르지면 결과의 값은 동일합니다

```python
# groupby의 다중 그룹화와 값은 동일하게 생성
group = df_gp.groupby(['시도', '진료년도'])[['환자수','요양급여비용총액']].sum()
group
```

### *corr()* 를 활용해서 컬럼간의 상관관계 파악하기
`corr()` 는 변수간의 유사율을 상관계수로 연산하여 표현 합니다.

- 1에 가까울 수록 비슷한 경향을 가짐
- -1에 가까울 수록 반대의 경향을 가짐
- 0에 가까울 수록 변수간의 상관 관계가 적음

```python
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원.csv',
                  encoding='CP949',
                  index_col=2,
                  header=1)
```

`method` 인자에는 상관관계를 연산하기 위한 알고리즘을 입력합니다.

- `pearson`: 피어슨 상관 계수 (선형 비교) - 기본값
- `kendall` : 켄달타우 상관 계수 (순위 비교)
- `spearman` : 스피어만 상관 계수 (순위 비교)

```python
cols = ['보험자부담금','요양급여비용총액','환자수','명세서청구건수','입내원일수']
corr = df_gp[cols].corr(method='pearson')
corr
```

> 값이 높으면 전반적인 변화의 경향성이 유사 (강남구 -> 강동구 부담금 감소시 급여비총액 감소 확률 높음)

```python
cols = ['보험자부담금','요양급여비용총액','환자수','명세서청구건수','입내원일수']
corr = df_gp[cols].corr(method='kendall')
corr
```

> 값이 높으면 순위의 변화가 유사 (부담금의 순위가 높은 구가 급여비총액 순위가 높을 확률 높음)

## 8. 데이터 결합
데이터의 파일이 1개로 통합되어 있지 않고 2개 이상으로 나누어져 있는경우 최종적으로 1개의 DataFrame으로 통합해야 분석이 용이한 경우가 많습니다.  

pandas에서는 크게 2가지의 데이터 결합방식을 제공합니다.

- **concat** : 2개 이상의 DataFrame을 구조적으로 연결
- **merge** : 2개의 DataFrame을 동일 컬럼을 기준으로 병합


### *concat( )* 을 활용한 구조적 결합(연결)

`concat()` 함수는 기본 인자로 입력된 여러개의 DataFrame 목록(리스트)를 특정 축(axis)로 연결하여 하나의 DataFrame을 반환하는 연결 함수 입니다.

<center><img src="https://drive.google.com/thumbnail?id=11kPbcBmkG_xDK8v_9CiRvmYiWoqDEW5U&sz=w1000" width="600"/></center>


- 컬럼 또는 로우의 구조가 같은 여러개의 DataFrame을 연결
- 컬럼의 구조가 같으면 결합시 수직으로 연결(기본)
- 로우의 구조가 같으면 결합시 수평으로 연결(axis = 1)

```python
import pandas as pd
```

```python
# 행 결합
df1 = pd.read_csv('data/진료비/건강보험심사평가원_진료비_2020.csv', encoding='cp949')
df1
```

```python
df2 = pd.read_csv('data/진료비/건강보험심사평가원_진료비_2021.csv', encoding='cp949')
df2
```

```python
df3 = pd.read_csv('data/진료비/건강보험심사평가원_진료비_2022.csv', encoding='cp949')
df3
```

```python
# concat() 사용하여 순서대로 데이터 연결
df_cat = pd.concat([df1, df2, df3], ignore_index=True)
df_cat
```

> `ignore_index` : `True` 이면 기존 데이터프레임의 인덱스를 무시하고 초기화 `False` 이면 인덱스를 유지

### *glob()*을 활용한 여러개의 파일 일괄 연결

**pathlib** 라이브러리의 **Path** 객체와 *glob()* 함수를 활용하면 여러개의 파일 경로를 목록으로 가져올 수 있습니다. 이후 반복문을 통해 경로로 부터 DataFrame을 불러오면서 리스트를 만들면 한번에 `concat` 처리가 가능합니다.

`Path()` 객체 생성시 인자에 데이터파일이 있는 폴더 경로를 입력합니다.

```python
# pathlib 활용
from pathlib import Path
import pandas as pd

file_dir = Path('./data/진료비')
file_dir
```

생성된 `Path` 인스턴스에서 `glob()` 함수를 사용합니다. `glob()`의 기본 인자에는 `*` 기호가 포함된 문자열을 만들어 다양한 파일명을 가진 파일경로를 한번에 불러옵니다.

결과는 `generater` 형식으로 반환되기에 반드시 for문을 사용하여 실제 경로를 가져와야 합니다.

```python
# glob 함수를 이용하여 경로를 포함한 파일 이름 접근

file_list  = file_dir.glob('*.csv')
df_list = []
for path in file_list:
    # .name 변수를 통해 파일명 확인 가능
    print(f'{path.name} 파일 로드')
    df = pd.read_csv(path, encoding='cp949')
    df_list.append(df)

df_cat = pd.concat(df_list, ignore_index = True)
df_cat
```

> '*.csv': csv 확장자를 가지는 모든 파일 (.csv로 끝나는 모든 파일명)

### merge()를 활용한 레이블 조합(병합)
<br>
`merge()` 함수는 입력된 두 DataFrame에서 **동일한 컬럼(key)**를 기준으로 수평적으로 DataFrame을 병합하는 함수입니다.

<center><img src="https://drive.google.com/thumbnail?id=18NggOJxx3tDYIRU3mKtxl1UyEZqnjiET&sz=w1000" width="600"/></center>


- SQL의 JOIN, Excel의 VLOOKUP()과 같은 역할
- 동일한 컬럼(key)은 관측치의 고유한 값을 나타내는것이 좋음
- 같은 key를 가지는 다른 변수들이 옆으로 결합되기에 수평적으로 연결



```python
import pandas as pd
# 예제 데이터 불러오기
df_left  = pd.read_csv('data/고객 샘플.csv', encoding='cp949')
df_right = pd.read_csv('data/진료 샘플.csv', encoding='cp949')
```

```python
df_left
```

```python
df_right
```

- 첫번째 인자로 입력된 DataFrame은 왼쪽, 두번째 인자의 DataFrame은 오른쪽 위치에 배치
- `on` 인자에 입력하며 동일한 컬럼(key)를 입력

```python
pd.merge(df_left, df_right, on='고객번호')
```

<br>

#### how옵션을 통한 병합 방법 설정

두 DataFrame이 동일한 컬럼을 가지고 있더라도, 내부 항목이 완전히 일치하게 존재하지 않을 수 있습니다. 이때 `how` 옵션을 활용해서 다음과 같은 데이터 결합 방법 방법을 사용할 수 있습니다.

+ `inner`: inner join. 양쪽다 존재하는 항목만 포함 - 기본값
+ `left`:  left join. 왼쪽을 기준으로 존재하는 항목만 포함
+ `right`: right join. 오른쪽 기준으로 존재하는 항목만 포함
+ `outer`: full outer join. 모든 항목을 다 포함

```python
# merge(왼쪽테이블, 오른쪽테이블, how = 결합방식, on = 기준 컬럼)
# inner join
pd.merge(df_left, df_right, how='inner', on='고객번호')
```

```python
# left join
pd.merge(df_left, df_right, how='left', on='고객번호')
```

```python
# right join
pd.merge(df_left, df_right, how='right', on='고객번호')
```

```python
# full outer join
pd.merge(df_left, df_right, how='outer', on='고객번호')
```
