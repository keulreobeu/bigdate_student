#### ì „ì²´ ë°ì´í„°ì— í´ë Œì§• ì ìš©í•˜ê¸°

```python
import re

stopwords = ['','ì´ˆ','ì´','ë“±']

def cleaning(review):
    review = re.sub(r'[^\w\s]', '', review)  # íŠ¹ìˆ˜ë¬¸ì ì œê±°
    review = re.sub(r'[\n\t]', ' ', review)  # ì¤„ë°”ê¿ˆ, íƒ­ ì œê±°
    review = re.sub(r'\s+', ' ', review)  # ì—°ì†ëœ ê³µë°± ì œê±°
    review= review.strip() # ë¬¸ì¥ ì–‘ë ê³µë°± ì œê±°

    # ë¶ˆìš©ì–´ ì œê±°
    review = ' '.join([word for word in review.split() if word not in stopwords])
    # í•œê¸€ì ì œê±°
    review = ' '.join([word for word in review.split() if len(word) > 1])

    return review

# í´ë Œì§• í•¨ìˆ˜ ì ìš©
df['clean'] = df['text'].apply(cleaning)
df
```

```python
# íŒë‹¤ìŠ¤ ì¶œë ¥ ì˜µì…˜ ì„¤ì •(ë¬´í•œëŒ€ ì¶œë ¥)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def show_tag(text):
    mc_text = mecab.pos(text)
    kw_text = [(t.form, t.tag) for t in kw.tokenize(text)]
    max_len = max([len(mc_text),len(kw_text)])

    mc_text += [None for i in range(max_len - len(mc_text))]
    kw_text += [None for i in range(max_len - len(kw_text))]

    text_set = {
        'Mecab':mc_text,
        'Kiwi':kw_text
    }
    return pd.DataFrame.from_dict(text_set)
```

### ì „ì²´ë°ì´í„°ë¥¼ í’ˆì‚¬ íƒœê·¸ë¡œ í´ë Œì§•

```python
from mecab import MeCab

store_df = pd.read_csv('data/sentence2_test.csv').dropna()

mecab = MeCab()

# ì›í•˜ëŠ” í’ˆì‚¬ íƒœê·¸ ëª©ë¡ ì„¤ì •
use_tags = ['NNP', 'NNG', 'NP', 'IC', 'MAG']
# ë¶ˆìš©ì–´ ëª©ë¡
stopwords = ['ì˜', 'ì•ˆ', 'ë„ˆë¬´', 'ë”', 'ë˜', 'ã…ã…']

# í† í°í™” í•¨ìˆ˜ ì •ì˜
def cleaning(review):
    # í’ˆì‚¬ íƒœê·¸ì™€ í•¨ê»˜ í† í°í™”
    kwords = mecab.pos(review)

    # ì›í•˜ëŠ” í’ˆì‚¬ë§Œ í•„í„°ë§
    kwords = [w[0]for w in kwords if w[1] in use_tags]
    # ë¶ˆìš©ì–´ ì œê±° ì¶”ê°€
    kwords = [word for word in kwords if word not in stopwords]
    clean = ' '.join(kwords)

    return clean

# í´ë Œì§• í•¨ìˆ˜ ì ìš©
store_df['text_tag'] = store_df['text'].apply(cleaning)
store_df.to_csv('data/sentence2_tag_.csv',index = False)
store_df
```

#### sentencepiece ëª¨ë¸ì„ í™œìš©í•˜ì—¬ í† í°í™”
sentencepiece ëª¨ë¸ì´ ë§Œë“¤ì–´ì§„ í›„ `SentencePieceProcessor` ê°ì²´ë¥¼ í†µí•´ ìƒì„±ëœ ëª¨ë¸ì„ ë‹¤ì‹œ í”„ë¡œê·¸ë¨ì— ë¡œë“œë¥¼ í•©ë‹ˆë‹¤. ì´í›„ `encode()` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í† í°í™” í•©ë‹ˆë‹¤.

```python
import sentencepiece as spm
import pandas as pd
import re

sp = spm.SentencePieceProcessor(model_file='spm2000.model') # ëª¨ë¸ ë¡œë“œ
df = pd.read_csv('data/sentence2_test.csv')
text = df.loc[0,'text']
sp_list = sp.encode(text, out_type=str)
print(text)
print(sp_list)
```

```python
def text_spm(text):
    text = str(text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'[\n\t]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text= text.strip()
    sp_list = sp.encode(text, out_type=str) # ë¬¸ì¥ì„ í† í°í™”
    sp_text = ' '.join(sp_list)
    return sp_text

df['text_spm'] = df['text'].apply(text_spm)
df
```

```python
import re
import emoji
from mecab import MeCab

mecab = MeCab()

stopwords = ['','ì´ˆ','ì´','ë“±']

def cleaning(review):
    '''
    íŠ¹ìˆ˜ë¬¸ì, ì¤„ë°”ê¿ˆ, íƒ­ ì œê±°, ì—°ì†ëœ ê³µë°±, ë¬¸ì¥ ì–‘ë ê³µë°±,
    ë¶ˆìš©ì–´('','ì´ˆ','ì´','ë“±', ì/ëª¨ìŒ, ), ì´ëª¨ì§€ ì œê±°
    '''
    review = re.sub(r'[^\w\s]', '', review)  # íŠ¹ìˆ˜ë¬¸ì ì œê±°
    review = re.sub(r'[\n\t]', ' ', review)  # ì¤„ë°”ê¿ˆ, íƒ­ ì œê±°
    review = re.sub(r'\s+', ' ', review)  # ì—°ì†ëœ ê³µë°± ì œê±°
    review= review.strip() # ë¬¸ì¥ ì–‘ë ê³µë°± ì œê±°

    # ë¶ˆìš©ì–´ ì œê±°
    review = ' '.join([word for word in review.split() if word not in stopwords])
    # í•œê¸€ì ì œê±°
    review = ' '.join([word for word in review.split() if len(word) > 1])
    # ì´ëª¨ì§€ ì œê±°
    review = emoji.replace_emoji(review, replace='')
    # ã…‹, ã…, ã…  ê°™ì€ ììŒ, ëª¨ìŒ ì œê±°
    review = re.sub(r'[ã„±-ã…ã…-ã…£]+', '', review)

    pos = mecab.pos(review)
    words = [w for w, t in pos if not (t.startswith('J') or t in {'MAG', 'MAJ'})]

    cleaned = ' '.join(words)

    return cleaned

# í´ë Œì§• í•¨ìˆ˜ ì ìš©
df['clean'] = df['text'].apply(cleaning)
df
```
