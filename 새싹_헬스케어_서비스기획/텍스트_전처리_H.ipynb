## 텍스트 전처리 요소

텍스트 데이터는 데이터중에서도 가장 노이즈가 크며 변수가 많은 데이터입니다. 따라서 텍스트를 분석하기 앞서 분석의 용이성을 위해 몇가지의 전처리 작업을 해줍니다.

- **토큰화**: 텍스트를 의미 있는 작은 단위(단어, 구두점 등)인 토큰으로 분리하는 과정

- **벡터화**: 텍스트 데이터를 컴퓨터가 이해하고 처리할 수 있는 숫자 형태의 벡터로 변환하는 과정

### 숫자 인코딩

문자 토큰을 숫치형으로 나타내는 가장 기본적인 방법은 문자 토큰에 **고유숫자**를 부여 하는 방법입니다.   
이렇게 숫자 토큰으로 변환하는 과정을 **인코딩**, 역으로 변환하는 과정을 디코딩이라 합니다.   
기본적인 인코딩 방식은 모든 문자 토큰을 찾아 고유 번호를 맵핑시킨 **딕셔너리**를 만드는 방식입니다.

```python
## 고유 토큰을 찾기위해 set 구조 활용
uq = set()
for text in df['text']:
    sp = text.split(' ')
    for w in sp:
        uq.add(w)
uq
```

```python
## 고유 토큰에 고유 번호를 맵핑시킨 딕셔너리 구성
enc_dict = {}
for i,t in enumerate(uq):
    enc_dict[t] = i
enc_dict
```

```python
## 딕셔너리를 활용하여 인코딩
text = df.loc[0,'text']
str_tk = text.split(' ')
enc_tk = [enc_dict[w] for w in str_tk]
enc_tk
```

## 텍스트 클렌징

### 특수문자 제거하기
특수 문자 제거는 파이썬의 내장 모듈인 `re` 모듈을 import하여 `sub()` 함수를 이용해 줍니다.
> re.sub(수정될 문자열의 정규식, 수정후의 문자열, 문자열)

```python
## 특수문자 정규식을 첫번째 인자로 입력
## 공백을 두번째 인자로 입력
import re
rtext = re.sub(r'[^\w\s]', '', text)
rtext
```

> `r'[^\w\s]'` 정규식은 알파벳, 숫자(`\w`), 공백(`\s`)을 제외한 모든 특수문자를 찾음     
> 해당 특수문자를 빈 문자열(`''`)로 대체하여 제거

### 불용어 제거하기
불용어 제거는 파이썬의 문자열 함수와 리스트, 반복문등을 이용하여 구성 가능합니다.    
먼저 토큰화 이후 반복문을 통해서 해당 단어가 불용어에 속하는지 `if` `in` 문을 활용하여 체크하고 속하지 않으면 새로운 리스트에 담아 `join`해줍니다.

```python
rtext.split(' ')
```

```python
# 토큰화 후 반복문을 활용
sptext = rtext.split(' ')
st_words = ['','초','이','등']
nw = []
for w in sptext:
    if w not in st_words:
        nw.append(w)
rtext = ' '.join(nw)
rtext
```

#### 전체 데이터에 클렌징 적용하기

```python
import re

stopwords = ['','초','이','등']

def cleaning(review):
    review = re.sub(r'[^\w\s]', '', review)  # 특수문자 제거
    review = re.sub(r'[\n\t]', ' ', review)  # 줄바꿈, 탭 제거
    review = re.sub(r'\s+', ' ', review)  # 연속된 공백 제거
    review= review.strip() # 문장 양끝 공백 제거

    # 불용어 제거
    review = ' '.join([word for word in review.split() if word not in stopwords])
    # 한글자 제거
    review = ' '.join([word for word in review.split() if len(word) > 1])

    return review

# 클렌징 함수 적용
df['clean'] = df['text'].apply(cleaning)
df
```

> DataFrame 사용시 `apply()` 함수를 적극 활용

### 전체데이터를 품사 태그로 클렌징

```python
from mecab import MeCab

store_df = pd.read_csv('data/sentence2_test.csv').dropna()

mecab = MeCab()

# 원하는 품사 태그 목록 설정
use_tags = ['NNP', 'NNG', 'NP', 'IC', 'MAG']
# 불용어 목록
stopwords = ['잘', '안', '너무', '더', '또', 'ㅎㅎ']

# 토큰화 함수 정의
def cleaning(review):
    # 품사 태그와 함께 토큰화
    kwords = mecab.pos(review)

    # 원하는 품사만 필터링
    kwords = [w[0]for w in kwords if w[1] in use_tags]
    # 불용어 제거 추가
    kwords = [word for word in kwords if word not in stopwords]
    clean = ' '.join(kwords)

    return clean

# 클렌징 함수 적용
store_df['text_tag'] = store_df['text'].apply(cleaning)
store_df.to_csv('data/sentence2_tag_.csv',index = False)
store_df
```

## 바이트페어 변환

형태소 분석의 단점중 하나는 토큰의 기준이 이미 정해져 있다는 것입니다.    
바이트페어 변환은 텍스트를 **바이트단위(문자)** 로 묶어가며 빈도 확률에 따라 최적화된 토큰셋을 만들어 냅니다.    
따라서 형태소 분석과 다르게 가지고 있는 텍스트뭉치에 따라 토큰이 최적화 되어 나온다는 장점이 있습니다.    

### spm 모듈 활용
`sentencepiece` 모듈은 바이트페어 인코딩 기능이 있는 대표적인 파이썬 라이브러리입니다.   
sentencepiece 모듈을 활용하여 다음과정을 통해 토큰화가 가능합니다.
1. sentencepiece 모듈에서 SentencePieceTrainer 객체를 통해 텍스트뭉치에서 토큰셋(model) 파일을 만들어냅니다.   
2. sentencepiece 모듈의 SentencePieceProcessor객체에 생성한 토큰셋 파일을 변환기(인코더)를 만들어 냅니다.
3. 만든 인코더에서 encode() 함수를 활용하여 텍스트를 토큰화 합니다.

[sentencepiece](https://github.com/google/sentencepiece)

#### 텍스트 뭉치 데이터를 텍스트 파일로 생성

```python
import sentencepiece as spm
import pandas as pd
import re

df = pd.read_csv('data/sentence2_test.csv')

# 추가 쓰기모드('a')로 텍스트 파일 열기
with open('data/sentence2_test.txt', 'a', encoding='utf-8') as af:
    for text in df['text']:
        text = str(text)
        text = re.sub(r'[^\w\s]', '', text)  # 특수문자 제거
        text = re.sub(r'[\n\t]', ' ', text)  # 줄바꿈, 탭 제거
        text = re.sub(r'\s+', ' ', text)  # 연속된 공백 제거
        text= text.strip() # 문장 양끝 공백 제거
        try:
            af.write(text+'\n')
        except:
                pass

```

#### 텍스트 파일로 부터 토큰 사전 및 sentencepiece 모델 생성
`SentencePieceTrainer`의 `train()` 함수를 이용하여 sentencepiece 모델을 생성할 수 있습니다.

```python
spm.SentencePieceTrainer.train(input='data/sentence2_test.txt', # 텍스트 뭉치 파일
                            model_prefix='spm2000',       # 출력 모델 파일 이름
                            vocab_size=2000                       # 토큰 개수
                            )
```

#### sentencepiece 모델을 활용하여 토큰화
sentencepiece 모델이 만들어진 후 `SentencePieceProcessor` 객체를 통해 생성된 모델을 다시 프로그램에 로드를 합니다. 이후 `encode()` 함수를 사용하여 토큰화 합니다.

```python
import sentencepiece as spm
import pandas as pd
import re

sp = spm.SentencePieceProcessor(model_file='spm2000.model') # 모델 로드
df = pd.read_csv('data/sentence2_test.csv')
text = df.loc[0,'text']
sp_list = sp.encode(text, out_type=str)
print(text)
print(sp_list)
```

```python
def text_spm(text):
    text = str(text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'[\n\t]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text= text.strip()
    sp_list = sp.encode(text, out_type=str) # 문장을 토큰화
    sp_text = ' '.join(sp_list)
    return sp_text

df['text_spm'] = df['text'].apply(text_spm)
df
```

> spm를 활용한 바이트페어 토큰화 결과중 `_`는 기존 문장의 띄어쓰기를 의미합니다. 따라서 띄어쓰기로 join 하여도 기존 문장의 띄어쓰기 정보를 잃어버리지 않습니다.

## 벡터화

텍스트 백터화란 자연어로 구성된 텍스트를 숫자 벡터로 변환하여 컴퓨터가 이해하고 처리할 수 있도록 만드는 과정입니다.   
텍스트 데이터를 고정된 길이의 벡터로 표현함으로써, 통계적 분석 또는 머신러닝 모델에서 활용하기에 반드시 필요한 전처리 작업입니다.

**백터화 프로세스**
백터화는 기본적으로 토큰화와 인코딩 이후 진행됩니다.   
<center><img src="https://drive.google.com/thumbnail?id=1xl4xyc_6wWxVIFQRmmeMbE3rqRSJGJ5T&sz=w1000" width="400"/></center>



- Bag-of-Words(BoW)
    * 단어의 출현 여부나 횟수만을 기준으로 텍스트를 벡터화함.
    * 문맥 정보를 반영하기 어려운 단점이 있으나 구현이 간단하고 직관적.

- TF-IDF(Term Frequency-Inverse Document Frequency)
    * 단어 빈도(TF)와 전체 문서에서의 희소성(IDF)을 결합하여 단어의 중요도를 측정.
    * 비교적 BoW보다 문서 내에서의 의미 있는 단어를 잘 식별 가능.

- Word Embedding
    * Word2Vec, GloVe 등 기법을 통해 각 단어를 저차원 벡터로 표현.
    * 단어 간의 의미적 유사성을 벡터 공간에서 반영 가능.

- Sentence Embedding
    * BERT, Transformer 등 사전 학습된 딥러닝 모델을 활용하여 문장 전체를 벡터로 표현.
    * 단어 수준을 넘어 문장 단위의 의미를 포착 가능.

임베딩 기법은 다음 챕터에서 배워보도록 하겠습니다.

### 빈도 벡터화 (Bag-of-Words)

특정 문장내에서 특정 토큰의 빈도수로 문장을 표현하는 벡터입니다.   

<center><img src="https://drive.google.com/thumbnail?id=1wU--TaGn3mtrzloU_lRGJ3BQVJtrCHff&sz=w1000" width="400"/></center>

`sklearn` 모듈의 `CountVectorizer` 모델을 이용하여 간단하게 구현할 수 있습니다.
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

df = pd.read_csv('data/sentence2_tag.csv').dropna()

## max_df : 해당 확률 이상으로 빈도가 많은 토큰을 제외
## min_df: 해당 확률 이하로 빈도가 적은 토큰을 제외
cvec = CountVectorizer(max_df=0.9, min_df=10)
cvec.fit(df['text_tag']) # 토큰생성 (띄어쓰기 기반)
names = cvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```

```python
X = cvec.transform(df.loc[:100,'text_tag']) # 인코딩
x = X.toarray()
x.shape
```

```python
import numpy as np

csum = np.sum(x, axis = 0)
ind = names['선물']
csum[ind]
```

### TF-IDF 벡터화

TF-IDF라는 수치는 단어의 중요도를 나타내는 수치중 하나입니다.   
단어의 빈도인 TF에 단어의 역 문서 빈도인 IDF를 곱한값 입니다.    
이런 TF-IDF 값으로 문장을 벡터로 나타내는것이 TF-IDF 벡터화 입니다.


TF-IDF 벡터는 `sklearn` 모듈의 `TfidfVectorizer` 모델을 이용하여 간단하게 구현할 수 있습니다.   
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html#sklearn.feature_extraction.text.TfidfVectorizer)

```python
from sklearn.feature_extraction.text import TfidfVectorizer #tfid

tfvec = TfidfVectorizer(max_df=0.9, min_df=10)
tfvec.fit(df['text_tag']) # 토큰생성 (띄어쓰기 기반)
names = tfvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```

```python
X = tfvec.transform(df.loc[:5,'text_tag']) # 인코딩
x = X.toarray()
print(x.shape)
print(x[0,:200])
```

```python
x[3,705]
```

```python
import numpy as np

csum = np.mean(x, axis = 0)
ind = names['선물']
csum[ind]
```

nreview_mask.csv 파일의 리뷰 텍스트에서 다음 전처리를 진행

1. 공백 및 특수 문자제거
2. 형태소 분석 및 불용어 제거를 통해 토큰만으로 구성된 텍스트를 clean 컬럼에 할당
4. 3번 결과의 데이터프레임을 pre_text.csv로 저장
5. clean 텍스트를 빈도 또는 tf-idf 벡터로 변환하여 출력

```python
df = pd.read_csv('/content/data/nreview_mask.csv')
```

```python
import re
import emoji
from mecab import MeCab

mecab = MeCab()

stopwords = ['','초','이','등']

def cleaning(review):
    '''
    특수문자, 줄바꿈, 탭 제거, 연속된 공백, 문장 양끝 공백,
    불용어('','초','이','등', 자/모음, ), 이모지 제거
    '''
    review = re.sub(r'[^\w\s]', '', review)  # 특수문자 제거
    review = re.sub(r'[\n\t]', ' ', review)  # 줄바꿈, 탭 제거
    review = re.sub(r'\s+', ' ', review)  # 연속된 공백 제거
    review= review.strip() # 문장 양끝 공백 제거

    # 불용어 제거
    review = ' '.join([word for word in review.split() if word not in stopwords])
    # 한글자 제거
    review = ' '.join([word for word in review.split() if len(word) > 1])
    # 이모지 제거
    review = emoji.replace_emoji(review, replace='')
    # ㅋ, ㅎ, ㅠ 같은 자음, 모음 제거
    review = re.sub(r'[ㄱ-ㅎㅏ-ㅣ]+', '', review)

    pos = mecab.pos(review)
    words = [w for w, t in pos if not (t.startswith('J') or t in {'MAG', 'MAJ'})]

    cleaned = ' '.join(words)

    return cleaned

# 클렌징 함수 적용
df['clean'] = df['text'].apply(cleaning)
df
```

```python
df.loc[3,'clean']
```

```python
df.to_csv('pre_text.csv', index=False, encoding='utf-8-sig')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer #tfid

tfvec = TfidfVectorizer(max_df=0.9, min_df=10)
tfvec.fit(df['clean']) # 토큰생성 (띄어쓰기 기반)
names = tfvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```
