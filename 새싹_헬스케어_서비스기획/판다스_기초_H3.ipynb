## 표(테이블)형식과 Pandas

**표(테이블)형** 데이터는 일상과 실무에서 가장 많이 활용되는 데이터의 표면적 구조로  엑셀 파일, SQL 등 을 생각할 수 있습니다.
기본적으로 표형식 데이터는 데이터를 **행**과 **열**로 나누어 **관측치**를 행으로, **변수**를 열로 배치하여 데이터를 보관합니다.

- **관측치(Observation)**: 하나의 실험·측정·사례에 해당하는 한 줄(row)의 데이터

- **변수(Variable)**: 데이터셋의 특성·속성에 해당하며, 관측치마다 값을 가지는 열

**pandas**는 **표(테이블)형** 데이터를 다양한 처리방식과 통계적인 기법으로 다룰수 있는 라이브러리로, 데이터분석에 있어 가장 많이 활용되는 대표적인 파이썬 모듈입니다.


### pandas의 데이터구조

**pandas**는 **DataFrame**과 **Series**라는 형식을 활용해서 데이터 구조를 만들고 다양한 연산과 처리를 지원합니다.

- `DataFrame`: **row(행, 관측치)** 와 **column(열, 변수)** 로 표현되는 라벨(이름)을 가진 2차원 표(테이블)형 데이터 구조.
    - row(행, 관측치)의 라벨(이름)은 인덱스(index)라고 표현
- `Series`: 동일 타입의 여러 개 데이터를 **라벨(index)**와 함께 저장하는 1차원 배열형 데이터 구조.

- `DataFrame`에서 하나의 열 또는 행을 선택하면 `Series`로 표현.
- 표(테이블) 형태를 가지는 데이터베이스(SQL)나 엑셀 데이터는 바로 `DataFrame`으로 가져와 분석이 가능.

### DataFrame생성 과 특징
`DataFrame`은 기본적으로 `DataFrame()` 클래스 생성자를 활용하여 생성 합니다.

이때, 생성자 인자에 `DataFrame`으로 나타낼 2차원(딕셔너리내의 리스트, 리스트내의 딕셔너리, 2차원 배열, 또다른 데이터프래임 등) 구조를 입력해 줍니다.


```python
# 라이브러리 불러오기
import pandas as pd
```

계층화된 딕셔너리구조를 이용하면 기본적으로 **키**가 **컬럼**이름으로 배치가 됩니다.

```python
# 딕셔너리를 활용한 DataFrame 생성
df = pd.DataFrame({'FIRST' : ['A', 'B', 'C', 'D'],
                   'SECOND': [7,6,5,8],
                   'THIRD' : ['2023','2022','2021','2020']})
df
```

```python
df = pd.DataFrame([{'FIRST':'A','SECOND':7,'THIRD': 2023},
                   {'FIRST':'B','SECOND':6,'THIRD': 2022},
                   {'FIRST':'C','SECOND':5,'THIRD': 2021},
                   {'FIRST':'D','SECOND':8,'THIRD': 2020}])
df
```

추가 인자를 활용하여 정보를 입력할 수 있습니다.
- `columns` 인자로 컬럼이름 초기화    
- `index` 인자로 로우이름(인덱스) 초기화

```python
# 2차원 구조이면 생성 가능
df = pd.DataFrame([[7,'2023','남자'],
                   [6,'2022','여자'],
                   [5,'2021','여자'],
                   [8,'2020','남자']],
                  index = ['A','B','C','D'],
                  columns = ['FIRST', 'SECOND', 'THIRD'])
df
```

## 2. 파일 불러오기

pandas는 빅데이터를 분석하기 위해 활용되기 때문에, 데이터를 직접 생성해서 만들기본단 기존에 가지고있는 규모가 있는 데이터 파일로 부터 `DataFrame`을 만드는 과정이 더욱 빈번합니다.

다음과 같이 pandas는 다양한 형식의 파일로 부터 `DataFrame`으로 데이터를 불러오는 함수를 제공합니다.

- `pd.read_csv()`: 콤마로 구분된 파일(.csv)을 읽어오는 함수
- `pd.read_excel()`: 엑셀 파일(.xls, .xlsx)을 읽어오는 함수
- `pd.read_json()`: JSON 형식의 파일을 읽어오는 함수
- `pd.read_html()`: HTML 파일에서 테이블을 추출하여 데이터프레임으로 변환하는 함수
- `pd.read_sql()`: SQL 쿼리 결과를 데이터프레임으로 읽어오는 함수
- `pd.read_spss()`: SPSS의 .sav 파일을 읽어오는 함수

#### Encoding 지정
pandas는 파일을 읽어올때 기본적으로 문자의 유니코드를 utf-8로 읽어옵니다.
하지만 파일이 사용하는 문자의 유니코드 형식이 utf-8가 아닌경우 `encoding` 인자에 파일의 코드형식을 입력합니다.  
한국어는 주로 `cp949` 코드를 사용합니다.

```python
# 인코딩(글자 저장 방법) 오류 발생 - UnicodeDecodeError
pd.read_csv('data/건강보험심사평가원.csv')
```

```python
# 옵션 encoding='CP949' 추가
pd.read_csv('data/건강보험심사평가원.csv', encoding='CP949')
```

> `CP949` : 한글 인코딩 방식 중 하나로, 주로 Windows 환경에서 사용

#### 라벨(이름) 지정
파일에서 특정한 행과 열을 row 와 column 라벨로 지정하고 싶은 경우 `index_col`과 `header` 인자를 활용합니다.

- `index_col`: 입력한 순서에 해당하는 열을 Row의 라벨로 설정
- `header`: 입력한 순서에 해당하는 행을 Column의 라벨로 설정

```python
# index, header(column) 지정
pd.read_csv('data/건강보험심사평가원.csv',
            encoding='CP949',
            index_col=2,
            header=1)
```

```python
# 리스트를 입력하여 다수열의 index또는 column 지정
pd.read_csv('data/건강보험심사평가원.csv',
            encoding='CP949',
            index_col=[1,2],
            header=1)
```

일반적인 비즈니스 데이터 분석에서 주제와 기간, 사이트, 제품, 공정 등 본인의 업무와 관련이 있는 일부 데이터만 선택하고 활용는 경우가 많습니다.

pandas의 DataFrame에서 기본적인 데이터선택 방법은 `[]` 인덱싱 연산을 활용하여 원하는 라벨의 데이터를 추출하는 방법입니다.



### 컬럼명을 활용한 변수 부분 선택

DataFrame은 `[ ]`연산에 **column(열/변수)** 이름을 넣어 데이터 해당 컬럼의 데이터들을 가져올 수 있습니다.

```python
# .columns 를 통해 컬럼이름 확인
df_gp.columns
```

## 조건을 활용한 인덱싱(필터링)
SQL에선 WHERE 절이나 Excel의 Filter와 같이 데이터에서 부분을 선택할 때 조건을 활용하는 경우 많습니다.
pandas를 활용한 데이터분석에도 유용한 데이터를 찾기위해 조건을 통해 원하는 데이터를 인덱싱해야할 경우가 많습니다.  

DataFrame에서도 조건을 활용하여 원하는 관측치를 찾아오는 필터링 인덱싱이 가능합니다.

```python
# 예제 데이터 불러오기
import pandas as pd
df_gp=pd.read_csv('data/건강보험심사평가원.csv',
                  encoding='CP949',
                  index_col=2,
                  header=1)

df_gp
```

### 데이터프레임 내부 데이터 정보 확인하기
조건을 통해 원하는 데이터를 찾기 위해서는 먼저 조건을 파악하기 위해 내부 데이터의 정보를 확인해야 합니다.

다음은 특정 변수에 대한 정보파악에 있어 기본적으로 활용되는 함수입니다.
- `unique()`: 범주형 시리즈에서 고유항목 출력   
- `min(), max()`: 수치형 시리즈에서 최소, 최댓값 출력

```python
# unique() 함수를 활용하여 특정 컬럼의 고유 항목을 가져올 수 있음
df_gp['시도'].unique()
```

```python
# nunique() 함수를 활용하여 특정 컬럼의 고유 항목을 개수를 가져옴
df_gp['시도'].nunique()
```

```python
# min(), max() 함수를 활용하여 데이터의 최대 최소값 확인 가능
print(df_gp['환자수'].min())
print(df_gp['환자수'].max())
```

### 시리즈 비교연산 & [bool] 인덱싱

조건을 활용한 필터링은 다음과 같은 두단계의 과정을 따릅니다.

- 원하는 컬럼을 선택하고 `비교연산자` 사용하여 bool 시리즈 생성

- `[bool 시리즈]`: 데이터프레임의 대괄호 연산에 bool 시리즈를 넣어 인덱싱   


```python
# 비교연산으로 bool 시리즈 생성
greater = df_gp['환자수'] > 379842
greater
```

```python
# bool 시리즈로 인덱싱
df_gp[greater]
```

`bool`타입을 가지는 시리즈끼리는 논리 연산자를 활용한 논리연산이 가능합니다.

- `~` : not   
- `&` : and    
- `|` : or    

```python
# 환자수가 100만 이상 True
greater = df_gp['환자수'] > 1000000
# 시도가 서울인경우 True
equal = df_gp['시도'] == '서울'

# 두 조건이 True인경우 True
filters = greater & equal

df_ft = df_gp[filters]
df_ft
```

논리 연산자를 연달아 사용할때는 연산자 간의 우선순위를 `()`기로호 묶어 고려해주어야 합니다.

```python
df_gp[(df_gp['환자수'] > 1000000) | (df_gp['시도'] == '서울')]
```

<br>

### 함수를 활용한 인덱싱
함수를 활용하여 bool 시리즈를 만들수도 있습니다.

- `isin()`: 인자로 입력한 항목에 해당되는 데이터를 True로 표현하는 bool 시리즈 반환   
- `between()`: 수치형에서 범위 내에 해당되는 데이터를 True로 표현하는 bool 시리즈 반환  

####  *isin()* 을 활용한 범주 인덱싱
여러개의 `==` 연산 결과에서 `|` 연산한 결과와 동일

```python
# isin()을 활용하여 포함여부로 bool값 산출
cond1 = df_gp['시도'].isin(['서울','경기','부산'])
cond1
```

```python
df_gp[cond1]
```

#### *between()* 을 활용한 수치 범위 인덱싱
`<` 과 `>` 의 연간 결과를 `&` 연산한 결과와 동일

```python
# between() 함수를 활용하여 하나의 컬럼(Series)에서 특정 범위에 대한 bool값 산출
df_gp['환자수'].between(300000, 1000000)
```

```python
# 양쪽 끝 경계 포함 여부 지정 가능b
df_gp[df_gp['환자수'].between(300000, 1000000, inclusive='left')]
```

> `inclusive` 인자에 `'both'`, `'left'`, `'right'` 중 하나를 입력하여 포함여부 결정

#### *sample()*  활용하여 랜덤한 표본 인덱싱
- `frac`는 백분율로 표본 추출
- `n`은 개수를 지정하여 표본 추출

```python
# sample()을 활용하여 임의의 표본 추출
df_gp.sample(frac=0.1)
```

```python
df_gp.sample(n=100)
```

## 데이터 변형 및 조작
pandas에는 DataFrame의 데이터를 변형하고 조작 하기 위한 다양한 기능들이 포함되어 있습니다.

이런 기능들은 주로 함수로 제공되며 모든 함수을 외우고 활용하기는 한계가 있어 문서 사이트를 통해 기능 함수를 찾아 활용할 수 있어야 합니다. (GPT등을 활용하여 더 빨리 찾기 가능)

**대표적인 정보처리 기능 함수**
- `rename()` : 인덱스, 컬럼 이름 수정

- `fillna()` : 결측값을 지정된 값으로 채움

### 데이터타입 변환
DataFrame에서는 컬럼별로 데이터타입을 가지기 때문에 데이터타입을 바꾸기 위해서는 먼저 특정 컬럼을 선택하여 시리즈를 가져와야 합니다.

특정 컬럼을 선택한 시리즈에서 `astype()`함수를 활용하면 바뀐 데이터타입의 시리즈를 얻을 수 있습니다.

```python
import pandas as pd
df = pd.DataFrame({'ID' : ['A', 'B', 'C', 'D','E'],
                   'Num': [4,2,3,5,1],
                   'Gender': ['Female', 'Male', 'Female', 'Male','Female'],
                   'Point': [2.1, 6.2, 3.4, 1.5, 7.8],
                   'Agree': [False, True, False, True, True],
                   'DATE' : ['2023-03-01','2022-07-08','2021-02-16','2020-09-23','2024-05-13']})
df
```

```python
# 각 컬럼별 데이터타입 확인
df.dtypes
```

> 항상 컬럼들의 데이터타입을 확인해주는것이 좋습니다.

#### *astype()* 시리즈 데이터타입 변환

- `astype()` :인자로 입력된 데이터타입을 가진 시리즈를 반환



```python
# 형변환 결과를 새로이 반환하므로 꼭 재할당 할것
df['NPoint'] = df['Point'].astype('int64')
df
```

> 타입의 변환은 데이터의 손실이 발생할 수 있으며(실수 -> 정수), 결측치가 존재시 타입변환 실패

```python
# 날짜 포맷의 문자열 또는 숫자형은 날짜형으로 변환가능
df['DATE'] = df['DATE'].astype('datetime64[ns]')
df.dtypes
```

> 년도-월-일 형식으로 표기된 문자열에서 astype() 으로 날짜형을 변환

### 정보 변환
다음은 DataFrame의 메타정보, 주로 인덱스와 컬럼정보를 수정하는 함수들 입니다.

```python
import pandas as pd
df = pd.DataFrame({'ID' : ['A', 'B', 'C', 'D','E'],
                   'Num': [4,2,3,5,1],
                   'Gender': ['Female', 'Male', 'Female', 'Male','Female'],
                   'Point': [2.1, 6.2, 3.4, 1.5, 7.8],
                   'Agree': [False, True, False, True, True],
                   'DATE' : ['2023-03-01','2022-07-08','2021-02-16','2020-09-23','2024-05-13']})
```

#### *rename()* 함수를 활용하여 인덱스, 컬럼의 이름 수정

- 키(수정할 이름)과 값(수정후 이름)이 매핑된 구조를 기본인자에 입력    
- `axis`: 변환 축을 선택 `0`이면 인덱스, `1`이면 컬럼

```python
# rename()를 이용한 인덱스 변환
df = df.rename({0:'A', 1:'B'}, axis=0)
df
```

```python
# rename()를 이용한 컬럼 이름 변환
df = df.rename({'Num' : 'Rank'}, axis = 1)
df
```

#### *set_index()* 를 활용해서 인덱스를 원하는 컬럼의 값으로 설정

- 인덱스로 바뀔 컬럼이름을 기본인자(`key`)로 설정

```python
# 인덱스로 바뀔 컬럼이름을 key 인자로 설정
df_si = df.set_index('ID')
df_si
```

#### *reindex()* 를 활용해서 기존 라벨을 원하는 순서로 재설정

- 원하는 인덱스 순서를 리스트로 만들어 기본 인자에 입력
- `axis` 인자로 변환 축 선택

```python
# 원하는 인덱스 순서를 넣어 변경
df_i = df_si.reindex(['E', 'B', 'C', 'A', 'D'], axis=0)
df_i
```

```python
# 원하는 컬럼 순서를 넣어 변경(axis=1)
df_c = df_i.reindex(['Gender', 'Agree', 'Point', 'Rank', 'DATE'], axis=1)
df_c
```

#### *reset_index()* 를 활용해서 인덱스를 순서로 초기화

```python
# 기존 인덱스는 컬럼으로 수정
df_ri = df_c.reset_index()
df_ri
```

### 데이터 변형
다음은 DataFrame 또는 Serise 내부의 데이터를 변형하는 다양한 함수들 입니다.

```python
df = pd.DataFrame({'ID' : ['A', 'B', 'C', 'D','E'],
                   'Num': [4,2,3,5,1],
                   'Gender': ['Female', 'Male', 'Female', 'Male','Female'],
                   'Point': [2.1, 6.2, 3.4, 1.5, 7.8],
                   'Agree': [False, True, False, True, True],
                   'DATE' : ['2023-03-01','2022-07-08','2021-02-16','2020-09-23','2021-02-16']})
```

#### *replace()* 를 활용해서 내부 데이터를 변환

- 키(변형할 데이터)와 값(변형될 데이터)가 매핑된 구조를 기본인자로 입력


```python
df['Gender'] = df['Gender'].replace({'Female': 'F', 'Male': 'M'})
df
```

#### *sort_values()* 를 활용해서 데이터를 정렬

- 정렬의 기준이 되는 컬럼이름을 기본 인자로 입력 (리스트로 여러개 입력가능)
- `ascending`: `False`이면 내림차순 정렬 `True`이면 오름차순 정렬

```python
# 나이순으로 내림차순 정렬
df_sort = df.sort_values('Point', ascending=False)
df_sort
```

### 결측치의 처리

결측치는 누락된 데이터 즉, 다른 관측치에는 있지만 특정 관측치에는 없는 데이터를 의미합니다. 이런 결측치는 분석에 있어 방해 요소가 되기때문에 적절한 방법으로 처리를 해주어야 합니다.

- 결측치는 DataFrame 내에서 Nan 또는 None으로 표기 됩니다.

<center><img src="https://drive.google.com/thumbnail?id=1DSMHTJ_whgzBUbI2XudAp4oA_M2EGK7X&sz=w1000" width="560"/></center>


```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'ID' : ['A', 'B', 'C', 'D','E'],
                   'Num': [4,2,3,5,1],
                   'Gender': [np.nan, 'Male', 'Female', 'Male','Female'],
                   'Point': [2.1, 6.2, 3.4, np.nan, 7.8],
                   'Agree': [False, True, np.nan, True, np.nan],
                   'DATE' : ['2023-03-01','2022-07-08','2021-02-16','2020-09-23','2021-02-16']})

df
```

#### *isna()* 로 결측치 유무 판단하기

- 함수 결과 결측치에 해당하는 부분만 `True`로 표기된 `DataFrame`이 만들어집니다.

```python
# isna() 함수를 활용하여 bool 값으로 나타내기
df.isna()
```

`any()`함수를 연달아 사용하여 결측이 있는 관측치 또는 컬럼을 찾을 수 있습니다

- `any()` : bool타입의 데이터프레임 또는 시리즈에서 특정 축에 하나의 `True`라도 있으면 `True`값을 반환

```python
# any() 함수로 결측치 유무 판단
df.isna().any(axis=1)
```

#### *count()*, *info()* 로 결측치 규모 파악

`count()` 함수는 결측을 제외한 데이터만 카운팅 하므로 결측치의 규모를 간접적으로 파악할 수 있습니다.

비슷하게 `info()` 함수결과에는 `count()` 정보가 포함됩니다.

```python
# count() 함수로 파악하기
df.count()
```

> 전체 데이터 개수보다 적으면 결측치 포함

```python
df.info()
```

#### *dropna()* 를 활용해서 결측치 삭제
결측치를 처리하는 방법중 하나는 결측치가 있는 관측치 자체를 삭제하는 방법입니다. 해당방법은 간편하지만 데이터의 손실이 크다는 단점을 가집니다.

- `subset` 인자에 결측을 확인할 컬럼 목록을 입력

```python
# subset 인자에 결측치를 찾을 컬럼입력(여러개는 리스트)
df_dn = df.dropna(subset=['Gender', 'Point'])
df_dn
```

> 결측이 존재하는 관측치 전부를 삭제하여 데이터 손실이 발생


#### *fillna()* 를 활용해서 NaN(결측치) 데이터 채우기
데이터의 손실을 최소화 하면서 결측치를 처리하는 방법은 결측치를 다른 데이터로 치환하는 방법입니다. 이때 분석에 왜곡을 최소화 할 수 있는 데이터로 치환해주는게 좋습니다.

```python
# fillna() 함수에 숫자를 넣어 해당 값으로 채움
df_fna = df.fillna(0)
df_fna
```

> 분석에 왜곡을 줄수 있음   

결측치를 처리할때 컬럼별로 특징이 다르기 때문에 컬럼 특징에 맞는 데이터로 각각 다르게 치환을 해주는것이 좋습니다.

`fillna()` 함수에 기본 인자로 컬럼-치환값 매핑 구조를 입력하면 컬럼별로 다른 데이터를 지정할 수 있습니다.

```python
# 맵핑 형태의 데이터로 컬럼별 원하는 값으로 채움
df_fna = df.fillna({'Gender': 'Male', 'Point' :1.4, 'Agree':True})
df_fna
```

#### *fillna()* 과 통계식을 활용해서 결측치 채우기

```python
# 수치형 컬럼에서 평균 산출
mean = df[['Point', 'Num']].mean()
mean
```

> 데이터 프레임에서 통계 함수 결과는 시리즈 이므로 인덱스-값 매핑형 구조를 가짐

```python
# 통계 연산이 컬럼이름과 매핑된 구조를 입력
df_fna = df.fillna(mean)
df_fna
```

```python
# 범주형 컬럼에서 최빈값 산출
mode = df['Agree'].mode()
print(mode)
```

```python
df_fna = df.fillna({'Agree':mode[0]})
df_fna
```

```python
# 연산 결과로 결측치 채우기
point = (6 - df['Num']) * 1.2
print(point)
```

> 시리즈에서 연산자 사용시 일괄연산 처리

```python
df_fna = df.fillna({'Point': point})
df_fna
```

`bfill()` 또는 `ffill()` 함수는 앞뒤 관측치와 동일한 값으로 결측치를 치환하기 때문에 유사한 데이터로 정렬된 상태에서 활용하기 좋습니다.




```python
# 다음 관측치의 값으로 결측치를 채움
df_fna = df.bfill() # 앞은 ffill
df_fna
```
