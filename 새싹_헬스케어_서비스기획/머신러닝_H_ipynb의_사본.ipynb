## 머신러닝 개요

탐색적 자료분석(EDA)는 데이터의 규모가 너무 크거나 비정형적인 데이터에서는 활용이 어렵다는 단점이 있습니다. 때문에, 다양한 변수에 따른 특정 변수의 값을 예측하는 추론의 영역에서는 EDA의 한계점이 분명하게 존재합니다.

탐색의 영역이 아닌 추론의 영역에서는 모델을 활용한 분석을 기반으로 합니다. 이때 학습을 통해 모델의 알고리즘(함수)를 만들어내는 방식을 **머신러닝(기계 학습)** 이라 합니다.

<center><img src="https://drive.google.com/thumbnail?id=1Ma1dcQS3XkbNobVmcjWavIpoLAilLWh5&sz=w1000" width="600"/></center>


머신러닝은 학습 방식과 문제 유형에 따라 다양한 모델 종류를 가지고 있습니다.

**학습 방식**
- **지도 학습**(Supervised Learning): 정답(라벨)이 있는 데이터셋을 학습하여 새로운 데이터를 예측.
- **비지도 학습**(Unsupervised Learning): 정답(라벨)이 없는 데이터에서 패턴을 찾아내는 방식 (예: 군집 분석).
- 강화 학습(Reinforcement Learning): 보상(Reward) 체계를 통해 최적의 행동을 학습. (예: 알파고, 자율주행)

**문제 유형**
- 회귀 모델 : 특정 값(수치)를 추론하는 모델
- 분류 모델 : 특정 범주(카테고리)를 추론하는 모델
- 군집화 모델 : 데이터의 특징으로 군집을 나누는 모델

# 회귀와 분류

**회귀 모델**은 **특정 값**의 추정치를 예측하는 모델로, 대부분은 라벨이 있는 데이터세트를 학습하는 지도학습 방식을 따릅니다.

반면에 **분류 모델**은 **특정 범주(클래스)**의 추정치를 예측한다는 차이가 있습니다.
분류 모델 또한 대부분 라벨이 있는 데이터세트를 학습하는 지도학습 방식을 따릅니다.   

## 1. 선형 회귀

$$y = Wx + b$$

선형 회귀는 위와 같이 흔히 사용되는 방정식을 만들어내는 모델입니다.  
W는 가중치, b는 편향을 나타냅니다.   
모델은 학습을 통해 해당 W와b를 유추하고 이에 따라 입력값(x)로 부터 결과값(y)를 추론 합니다.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('Healthcare-Diabetes.csv')
df = df[df['Insulin'] > 0]
````

````python
plt.scatter(x=df['Glucose'], y=df['Insulin'], s=10)
plt.show()
````

### sklearn의  linear_model

sklearn에서는 모델을 학습, 평가등의 기본적인 기능을 넣어 객체화 시켜놓았습니다. 따라서 모델을 사용하는데 있어 간단하게 객체만 불러오고 함수를 통하 학습 및 예측이 가능합니다.

선형회귀 모델은 `linear_model` 모듈의 `LinearRegression` 객체로 제공이 됩니다.   
[LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression)

````python
# 모델 불러오기
from sklearn import linear_model

reg = linear_model.LinearRegression()
````

### 모델 학습하기
선형 회귀 모델은 다음과 같은 공식이 최소가 되는 방향으로 가중치를 업데이트 합니다.    

$$(Wx - y)^2$$

이렇게 모델 학습에 있어 업데이트의 방향을 잡아주는 공식을 **손실 함수**라고 합니다.


````python
train_X = np.reshape(df['Glucose'].to_numpy(), (-1,1))
y = df['Insulin'].to_numpy()
train_X.shape
````

**fit() 함수로 학습**

sklearn의 대부분 모델은 `fit()` 함수를 이용하여 학습을 진행합니다.   
선형 회귀 모델은 지도학습이므로 첫번째 인자는 입력데이터(feature)를 두번째 인자는 결과데이터(label)를 입력합니다.  

````python
reg.fit(train_X, y)
````

````python
reg.coef_
````

### 모델 예측하기
LinearRegression 모델의 `predict()` 함수를 활용하여 입력데이터를 넣어 예측값을 가져올수 있습니다.  
이때 입력데이터는 학습데이터의 입력형태와 동일해야 합니다.

````python
df['Insulin_p'] = reg.predict(train_X)
df
````

````python
plt.scatter(x=df['Glucose'], y=df['Insulin'], s=10)
plt.plot(df['Glucose'], df['Insulin_p'], color='red')
plt.show()
````

### 회귀 모델 평가

모델의 가중치가 잘 형성되었는지 파악하기 위한 평가지표가 있습니다.   
회귀모델은 다음과 같은 평가지표를 활용하여 모델의 정합성을 판단합니다.

- mean_absolute_error(MAE): 평균 절대 오차값
- mean_squared_error(MSE): 평균 제곱 오차값
- mean_absolute_percentage_error(MAPE): 평균 절대 오차 확률값
- r2_score: 피어슨 상관계수의 제곱값 (변수 X 와 Y 간의 선형 상관 관계)

MSE, MAE등은 직관적인 지표가 아니므로 0 ~ 100 사이의 확률을 나타내는 MAPE또는 0 ~ 1 사이의 값을 나타대는 R2 스코어를 많이 활용합니다.

````python
from sklearn.metrics import mean_absolute_percentage_error, r2_score

mae = mean_absolute_percentage_error(df['Insulin'], df['Insulin_p'])
r2 = r2_score(df['Insulin'], df['Insulin_p'])
print(mae,r2)
````

## 1. 로지스틱 분류
로지스틱 모델은 선형 회귀 모델의 결과(Wx)에 로짓 함수를 적용한 모델로, 최종 결과값을 선형적인 값이 아닌 0과 1사이의 극단화된 값으로 변형합니다.

$$ 1 / (1 + exp(-Wx)) $$

<center><img src="https://drive.google.com/thumbnail?id=1PAXXTgSNx0DActb0kH9t89BhECE2FwFd&sz=w1000" width="300"/></center>

이처럼 로짓 함수를 결과에 적용하면 결과값이 0~1사이의 확률값으로 나타나게 되고
이를 분류의 결과로 활용할 수 있습니다.

    이진 분류: 모델의 예측 결과가 0에 가까우면 부정, 1에 가까우면 긍정
    다중 클래스 분류: 모델의 만든 여러개의 확률값중 가장 높은 확률을 가지는 아이템으로 분류

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df_hd = pd.read_csv('Healthcare-Diabetes.csv')
df_hd
````

#### 데이터 확인

````python
sns.scatterplot(data=df_hd, x= 'Glucose',y='BMI',hue='Outcome')
plt.show()
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_X = df_hd[['Glucose','BMI']].to_numpy()
train_X = scaler.fit_transform(train_X)
y = df_hd['Outcome'].to_numpy()
````

#### 로지스틱 모델 학습

분류 모델은 확률 예측값(p)을 통해 다음과 같은 손실 함수가 최소가 되는 방향으로 가중치를 업데이트합니다.
$$ -log(p) $$


로지스틱 모델은 `LogisticRegression` 객체로 제공되며 지도학습이므로 동일하게 `fit()` 함수에 첫번째 인자는 입력데이터(feature)를 두번째 인자는 결과데이터(label)를 입력합니다.  
이때 결과데이터는 0또는 1로 구성된 정수세트여야 합니다.

````python
from sklearn.linear_model import LogisticRegression

lreg = LogisticRegression(random_state=0)
lreg.fit(train_X, y)
````

````python
lreg.coef_
````

````python
lreg.classes_
````

#### 임의의 입력으로 학습된 모델 예측값을 확인

````python
test_X = np.random.rand(500,2)

pred = lreg.predict(test_X)
````

````python
plt.figure(figsize = (10,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
sns.scatterplot(x= train_X[:,0],y=train_X[:,1],hue=y, ax = ax1)
sns.scatterplot(x= test_X[:,0],y=test_X[:,1],hue=pred, ax = ax2)
plt.show()
````

#### 분류 모델 평가

분류 문제에서 평가지표는 다음과 같은 지표를 사용합니다.

- accuracy: 분류값을 맞춘 확률, 기본적인 지표로 사용

- f1_score: 정밀도와 재현율의 조화 평균, 클래스의 균형을 파악할 수 있음
    - `average` : 다중 클래스에서는 `'weighted'`를 이진 분류에서는 `'binary'`사용

[f1_score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html)

````python
from sklearn.metrics import accuracy_score, f1_score

# 학습 데이터에 대한 예측 정확도 비교
pred = lreg.predict(train_X)

# 정확도 지표 확인
print(accuracy_score(y, pred))

# f1_score 지표 확인
print(f1_score(y, pred, average='binary'))
````

# 머신러닝 프로세스

머신러닝 모델의 이점은 분석을 통해 변수간의 관계가 명확하게 보기 힘들어도 여러 변수들간의 관계를 학습하여 원하는 결론을 도출 할수 있다는 점입니다.

따라서 머신러닝 모델을 활용할땐 여러개의 변수를 활용하여 회귀 또는 분류 학습을 진행하며 이에따라 필요한 변수를 선정 하는 과정이 필요합니다,

뿐만아니라 학습 이전에 학습 효율을 위한 전처리 과정도 필요하며 가진 데이터가 가장 효율적인 결과를 만드는 모델을 선정하는 과정도 필요합니다.

**머신러닝 전체 프로세스**

- 입력데이터(feature)와 결과데이터(label)선정

- 데이터 전처리
    * 스케일링, 이상치&결측치 처리

- 데이터세트 구축
    * 학습 데이터와 평가 데이터 분리

- 모델 선정 및 학습

- 모델 평가와 반복 실험

#### 이상치 처리
이상치를 처리하는 방법은 다양하지만, 크게 두가지 방식이 많이 활용됩니다.

- 평균과 표준편차: 평균에서 멀리 떨어진 데이터를 이상치로 판단하여 제거 (분포에 영향을 받음)

- 백분위수: 고정된 퍼센트를 기준으로 이상치를 판단하여 제거 (분포와 상관 없음)

````python
# 평균과 표준편차
mean = df[seq_X].mean()
std_dev = df[seq_X].std()

# 이상치 기준
lower_bound = mean - 10 * std_dev
upper_bound = mean + 10 * std_dev

outlier = ((df[seq_X] < lower_bound) | (df[seq_X] > upper_bound)).any(axis=1)
~outlier
````

````python
# 1,3 사분위수 기준
Q1 = df[seq_X].quantile(0.25)
Q3 = df[seq_X].quantile(0.75)

# IQR 계산
IQR = Q3 - Q1
lower_bound = Q1 - 3.5 * IQR
upper_bound = Q3 + 3.5 * IQR

outlier = ((df[seq_X] < lower_bound) | (df[seq_X] > upper_bound)).any(axis=1)
~outlier
````

### 수치형 데이터 스케일링
연산을 통해 결과값을 산출하는 머신러닝 모델은 매개변수들의 수치 스케일이 차이가 크면 불안정한 추론을 가질 확률이 높아집니다.   
이를 방지하기위해 최소최대변환(MinMaxScaler) 또는 표준점수변환(StandardScaler)을 활용하여 수치의 스케일을 맞추는것이 좋습니다.

- `fit()`: 입력된 데이터에서 필요한 파라미터를 설정하는 단계 (최소,최대)    
- `transform()`: 입력된 데이터를 파라미터를 통해 변환 하는 단계 (스케일링)  

````python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[seq_X])
scale_X = scaler.transform(df[seq_X])
scale_X
````

### 데이터세트 구성

어떤 데이터를 활용할지 결정하였다면 최종적으로 회귀문제, 분류문제에 따라 입력변수(feature)-결과(lable)을 매핑한 최종 데이터구조를 만들어야 합니다.  
이렇게 모델의 형식에 맞게 정리된 데이터구조를 **데이터세트**라하며 주로 배열 형식으로 만들어줍니다.

#### 데이터세트 형상
- Feature 형상 : (샘플개수, 변수개수)
- label 형상: (샘플개수)

````python
# 입력데이터 선정
import numpy as np
import pandas as pd

# 스케일 결과와 인코딩 결과를 concat하여 입력데이터로 설정
X = np.concatenate((scale_X, enc_X), axis=1)

# 회귀, 분류에 대한 결과데이터 설정
RY = df[reg_Y].to_numpy()
CY = df[class_Y].to_numpy()

print(f'입력데이터 형상: {X.shape}')
print(f'회귀 결과데이터 형상: {RY.shape}')
print(f'분류 결과데이터 형상: {CY.shape}')
````

#### 학습 & 평가 분리
학습에 사용된 데이터를 이용하여 평가지표를 산출하면 **과적합** 모델이더라도 높은 평가지표를 얻게 될 것 입니다. 이는 객관성이 떨어지는 지표이며, 학습에 사용되지 않은 데이터를 입력한 평가지표여야 객관성을 얻을 수 있습니다.   

- 과적합 : 모델이 반복학습을 통해 학습데이터에만 맞추어져 학습되지 않은 데이터를 예측 못하는 경우

이렇게 학습에 사용되지 않고 모델의 최종 평가를 위한 데이터세트를 **평가 데이터세트**라고 합니다.  

대표적인 분리 방식은 sklean의 `train_test_split()` 함수를 이용하여 비율을 통해 랜덤하게 데이터세트 분리하여 학습세트와 평가세트를 나눕니다.
 - `test_size` : 평가 데이터의 비율    
 - `random_state` : 랜덤한 값을 만들때 필요한 시드값, 동일한 값을 주면 재실행 시에도 동일하게 데이터가 분리됨

````python
from sklearn.model_selection import train_test_split

train_X, test_X, train_RY, test_RY = train_test_split(X, RY,
                                                      test_size=0.2,
                                                      random_state=42)

print(train_X.shape, train_RY.shape)
print(test_X.shape, test_RY.shape)
````

### 회귀 모델 학습 및 평가

선형 회귀 모델은 학습을 위한 매개변수가 많을수록 다중공선성 문제가 발생하여 입력데이터 의 작은 변화에도 가중치의 변화가 커져 불규칙한 예측값을 나타낼 수 있습니다,
이러한 다중공선성 문제는 모델의 overfiting 문제의 하나의 원인이 되기도 합니다

**Ridge 회귀**
- Ridge 정규화를 추가하여 다음과 같은 공식으로 오차가 최소가 되는 방향으로 가중치를 업데이트
$$(Wx - y)^2 + \alpha W^2$$
- W의 제곱을 더함으로 W의 변동폭을 줄이고 안정적인 W값을 구성할 수 있어 다중공선성의 문제를 완화

**Bayesian 회귀**
- 가중치와 정규화 파라미터에 대해 확률적 관점을 적용하여 불확실성을 고려
- Ridge 회귀에 Bayesian 을 적용하여 적절한 alpha 값을 선정할 수 있음

[Ridge](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html#sklearn.linear_model.Ridge)   
[BayesianRidge](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.BayesianRidge.html#sklearn.linear_model.BayesianRidge)

- 모델의 입력데이터는 `NObeyesdad` 컬럼을 제외한 컬럼중 적절한 컬럼들을 선정
- 이상치 제거, 스케일링, 인코딩 등 필요에 따른 데이터 전처리 진행
- 학습 및 평가데이터를 분리하여 학습과 평가 진행
- 모델은 2가지 이상의 모델을 사용하여 f1스코어와 정확도를 비교
- SVC, SGDClassifier 등의 모델을 사용하는 경우 max_iter 인자를 설정하고 너무 큰 숫자를 입력하지 말것(오래걸림)
    * 예시: `SVC(C=100, max_iter=30)`

[분류 모델](https://scikit-learn.org/stable/supervised_learning.html)

````python
import pandas as pd
df1 = pd.read_csv('ObesityDataSet.csv')
df1.head()
````

````python
# 교통수단
# case1 보도 0, 대중교통 1, 개인 교통수단 2
# case2 원핫때려잇
print(df1['MTRANS'].unique())
````

````python
# Normal_Weight: 정상체중, Insufficient_Weight: 저체중,
# Overweight_Level_I: 과체중, Obesity_Type_I: 비만
# 저체중 0 정상1 과체중 3,4 비만 5,6

print(df1['NObeyesdad'].unique())
````

````python
# case1 대중교통 분류 적용
ml_df = df1.copy()

bool_list = ['family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']
cat_list = ['CAEC','CALC']

ml_df['Gender'] = ml_df['Gender'].replace({'Female':1, 'Male':0})
ml_df[bool_list] = ml_df[bool_list].replace({"yes":1, "no":0})
ml_df[cat_list] = ml_df[cat_list].replace({'Sometimes':1, 'Frequently':2, 'Always':3, 'no':0})
ml_df['MTRANS'] = ml_df['MTRANS'].replace({'Public_Transportation':1, 'Walking':0, 'Automobile':2, 'Motorbike':2, 'Bike':2})
ml_df['NObeyesdad'] = ml_df['NObeyesdad'].replace(
    {'Normal_Weight':1, 'Insufficient_Weight':0, 'Overweight_Level_I':2, 'Overweight_Level_II':3, 'Obesity_Type_I':4, 'Obesity_Type_II':5, 'Obesity_Type_III':6})
````

````python
ml_df.head()
````

````python
from sklearn.model_selection import train_test_split

X = ml_df.drop(['NObeyesdad'], axis=1)
y = ml_df['NObeyesdad']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
````

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier



models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'CatBoost': CatBoostClassifier(verbose=0)
}
````

````python
# 원핫, 스케일링x

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(f'{model_name} - Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}')
````

````python
models_opti = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(
        n_estimators=500,
        random_state=42,
        n_jobs=-1
    ),
    'CatBoost': CatBoostClassifier( loss_function="MultiClass",
        iterations=1000,
        depth=8,
        learning_rate=0.1,
        random_state=42,
        verbose=False,
        allow_writing_files=False)
}
````

````python
for model_name, model in models_opti.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(f'{model_name} - Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}')
````

1. 진료건수-기상청 파일을 로드하고 전처리 진행      
    - 일강수량(mm)의 결측치는 0으로 채우고 나머지 변수에 대한 결측치는 다음(밑)순서의 관측치 값으로 채우기(bfill)     
    - 날짜 컬럼으로 부터 요일 컬럼을 산출
        
2. '평균 상대습도(%)', '일강수량(mm)', '평균 풍속(m/s)', '최저기온(°C)', '최고기온(°C)', '요일' 변수를 통해 '감기' 변수를 예측하는 과정 만들기     
   - 마지막 600일을 평가 데이터로, 나머지 날짜를 훈련데이터로 나누기     
   - 입력 데이터는 표준점수 스케일링을 진행하고 예측 데이터는 평균으로 나누어 스케일링     
   - 서로 다른 2개의 회귀 모델을 사용하여 mape와 r2를 비교할것     
   - 학습 이후 평가데이터인 600일의 감기 예측값과 실제값을 비교하는 선그래프 그리기    

[회귀 모델](https://scikit-learn.org/stable/supervised_learning.html)

````python
df2 = pd.read_csv('/content/전료건수-기상청.csv', encoding='cp949')
````

````python
df2.head()
````

````python
reg_df = df2.copy()
reg_df.drop(['Unnamed: 0', '감기발생', '년도'], inplace=True, axis=1)
reg_df.drop(['천식','눈병','피부병'], inplace=True, axis=1)
reg_df['날짜'] = pd.to_datetime(reg_df['날짜'])
reg_df['일교차'] = reg_df['최고기온(°C)'] - reg_df['최저기온(°C)']
````

````python
reg_df.info()
````

````python
reg_df['요일'] = reg_df['날짜'].dt.dayofweek
reg_df['월달'] = reg_df['날짜'].dt.month
````

````python
reg_df['일강수량(mm)'] = reg_df['일강수량(mm)'].fillna(0)
reg_df = reg_df.fillna(method='bfill')
````

````python
reg_df.head()
````

````python
reg_df.drop(['날짜'], axis=1, inplace=True)
````

````python
X = reg_df.drop(['감기'], axis=1)
y = reg_df['감기']
````

````python
X['월달_sin'] = np.sin(2 * np.pi * X['월달']/12)
X['월달_cos'] = np.cos(2 * np.pi * X['월달']/12)

X['요일_sin'] = np.sin(2 * np.pi * X['요일']/7)
X['요일_cos'] = np.cos(2 * np.pi * X['요일']/7)
X.drop(['월달','요일'], axis=1, inplace=True)
````

````python
train_X = X.iloc[:-600]
train_y = y.iloc[:-600]
test_X = X.iloc[-600:]
test_y = y.iloc[-600:]

mean = y.mean()
train_y = train_y / mean
test_y = test_y / mean
````

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = ['평균기온(°C)','최저기온(°C)','최고기온(°C)','일강수량(mm)', '평균 풍속(m/s)', '평균 상대습도(%)', '일교차']

train_X.loc[:, cols] = scaler.fit_transform(train_X[cols])
test_X.loc[:,  cols] = scaler.transform(test_X[cols])
````

````python
train_X.describe()
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_absolute_error, mean_squared_error, median_absolute_error
from catboost import CatBoostRegressor

models = {
    'Linear Regression': LinearRegression(),
    'CatBoost': CatBoostRegressor(iterations=900,
                                  depth=7,
                                  loss_function='RMSE',
                                  eval_metric='RMSE',
                                  verbose=0,
                                  random_state=42)
}


for model_name, model in models.items():
    model.fit(train_X, train_y)
    pred_y = model.predict(test_X)
    mape = mean_absolute_percentage_error(test_y, pred_y)
    mse = mean_squared_error(test_y, pred_y)
    medae = median_absolute_error(test_y, pred_y)
    r2 = r2_score(test_y, pred_y)
    print(f'{model_name} - mape: {mape:.4f}, mse: {mse:.4f}, medae: {medae:.4f}, r2 Score: {r2:.4f}')
````
